RANDOM FOREST 

1. DATA EXPLORATION

```python
import pandas as pd

lung_cancer_detection = pd.read_csv("lung_cancer_detection.csv")
lung_cancer_detection.head()
```

```python
lung_cancer_detection.info()
```

2. DATA PREPARATION

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
lung_cancer_detection['GENDER'] = le.fit_transform(lung_cancer_detection['GENDER'])
lung_cancer_detection['LUNG_CANCER'] = le.fit_transform(lung_cancer_detection['LUNG_CANCER'])

y = lung_cancer_detection['LUNG_CANCER']
x = lung_cancer_detection.drop(columns=['LUNG_CANCER','GENDER'])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,y,
    train_size=0.7,
    stratify=y,
    random_state = 123
)
x_train.shape, x_test.shape
```

3. MODEL TRAINING AND EVALUATION

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

lung_cancer_detection_rf = RandomForestClassifier (
    n_estimators = 200,
    max_depth = None,
    min_samples_split = 2,
    min_samples_leaf = 1,
    bootstrap = True,
    random_state = 123
)

lung_cancer_detection_rf.fit(x_train, y_train)

y_pred = lung_cancer_detection_rf.predict(x_test)
accuracy_score(y_test,y_pred)
```
